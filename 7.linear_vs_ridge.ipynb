```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Use all features
X = diabetes.data
y = diabetes.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create Linear Regression object
lr = LinearRegression()

# Train the model using the training sets
lr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred_lr = lr.predict(X_test)

# The mean squared error
print('Linear Regression Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred_lr))
```

```python
# Create Ridge Regression object
ridge = Ridge(alpha=4.0)

# Train the model using the training sets
ridge.fit(X_train, y_train)

# Make predictions using the testing set
y_pred_ridge = ridge.predict(X_test)

# The mean squared error
print('Ridge Regression Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred_ridge))
```

```python
# Plot coefficients
plt.figure(figsize=(10, 6))
plt.plot(range(len(diabetes.feature_names)), lr.coef_, color='g', label='Linear Regression')
plt.plot(range(len(diabetes.feature_names)), ridge.coef_, color='b', label='Ridge Regression')
plt.xticks(range(len(diabetes.feature_names)), diabetes.feature_names, rotation=60)
plt.ylabel('Coefficients')
plt.legend()
plt.show()
```
