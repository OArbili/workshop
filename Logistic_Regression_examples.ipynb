```python
import numpy as np
import matplotlib.pyplot as plt

# Define the center points
center_points = np.array([[3, 0, 1], [1, 1, 0], [0, 3, 1]])

# Define the standard deviation for the noise
noise_std = 0.5

# Generate random data points around the center points with noise and labels
num_points = 100
data = []
labels = []
for center in center_points:
    x = np.random.normal(center[0], noise_std, num_points)
    y = np.random.normal(center[1], noise_std, num_points)
    label = np.full(num_points, center[2])
    data.append(np.vstack((x, y)).T)
    labels.append(label)

# Combine the data and labels from all center points
data = np.concatenate(data, axis=0)
labels = np.concatenate(labels, axis=0)

# Separate the data into x and y vectors
x = data[:, 0]
y = data[:, 1]

# Plot the scatter chart
plt.scatter(x, y, c=labels)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random Data with Labels')
plt.grid(True)
plt.show()
```
