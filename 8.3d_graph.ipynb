```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
x1 = [30, 45, 65, 80, 97, 60, 20, 13, 19]
x2 = [12, 14, 30, 45, 47, 38, 6, 3, 7]
y = [150, 123, 150, 178, 186, 168, 87, 60, 92]

# Create a larger figure
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(x1, x2, y,s=80,color='blue')

# Set labels and title
ax.set_xlabel('Weight',fontsize=16)
ax.set_ylabel('Age',fontsize=16)
ax.set_zlabel('Height',fontsize=16)
ax.set_title('Weight and Age vs Height',fontsize=30)

# Show the plot
plt.show()
```
